# Extracting the Text from a PDF File

```python
import pytesseract
from pdf2image import convert_from_path

# Path to the PDF file
pdf_path = "Concise-Islamic-Catechism.pdf"

# Convert PDF pages to images
pages = convert_from_path(pdf_path)

# Initialize an empty string to store all extracted text
all_text = ""

# Loop through each page
for i, page in enumerate(pages):
    # Save the image
    image_path = f"out/page_{i + 1}.jpg"
    page.save(image_path, "JPEG")

    # Extract text from the image
    extracted_text = pytesseract.image_to_string(page, lang="eng")
    all_text += f"Page {i + 1}:\n{extracted_text}\n"

# Write the extracted text to a text file
output_file = "Concise-Islamic-Catechism.txt"
with open(output_file, "w") as f:
    f.write(all_text)

print(f"Extracted text saved to {output_file}")
```

## FIRST CHAPTER

### BASES OF FAITH IN ISLAM


Islam is established on two principal bases of faith :

1. There is no other God worth worshipping except Allah,
2. That Muhammed (peace be upon him) is the prophet sent by God to all the humanbeings.

The religion of Islam has summed up as a printiple in this sentence :

لاَ إِلَٰهَ إِلَّا اللَّهُ مُحَمَّدٌ رَسُولُ اللَّهِ

**La Ilaahe illallah Muhammedun Rasulullah**

(Transliteration)

"There is no deity save Allah (God) and Muhammed is His servant and His apostle". (Translation).

The man, who confirms this holy sentence by word of mouth and certifies it with his heart, is considered as having entered the religion of Islam.

This Faith is called Iman (Belief) and one who believes it, is named as Mü’min (Believer).

Let us have a glimpse of this endless existing universe. The world we live in, is one amongst million parts of this universe. So far as human being is concerned, he is like a minute particle as compared with the greatness of the world. Structurally weak and puny, the man is however the most honourable and the highest creature of this existing world. No other creature except man could attain this eminence. The man who has attained this height and honour, has equivalent responsibilities at the same time.

Therefore it is necessary for every wise man to think about : 

From where he has come,
Why he has come,
And where he will go,

What do we mean by this world? How did heaven and this universe come into existence and how did they acquire the present shape?

And man is forced by his mind to find answers to these and the relevant questions. Everybody who is not sick and possesses a sound mind, asks
himself questions of this kind and by thinking, he is able to find the pertaining answers. Just starting from ourselves let us also think about the earth, heaven and the inhabitants and creatures present in them.

This world, endless and limitless skies with millions and billions of stars, suns and moons; as from when are they existing and until when will
they continue to exist? Was there any time when they were not existing and will the time come when they will be no more existing? Coming into this
world, growing, dying, moving about, coming to a halt, completion and breaking into small pieces, are all clear-cut indications that they come into existence later on. Present day knowledge which is based on experience and observation also accepts that this universe had come into existance at later stage.

The conclusion of our thought :

If it is so, then it becomes obvious that the earth, heaven and everything which is present on the earth and in heaven, were created later on.

Now let us have a look at the existing universe!

Every thing present in it can be divided into two principal sections :

1 — Living ones,
2 — Lifeless ones.

These living creations include human beings, animals and plants; and these lifeless ones include every thing not included in the first category. Humanbeing is at the top of the creations present in the first section. And man with the highest degree of wisdom, is dominant over animals, plants and matter. He flies in the skies and goes to the depths of the oceans, he can make the finest instruments and can make use of them, and he proves his superiority by bringing other creatures under his control by means of his wisdom.

Besides man’s superiority, not only one human-being but all humanity is not in a position to create even the most minute particle of sand out of nothing.

This universe has been created by God who does not belong or resemble to it.


RECOGNITION OF GOD

The creator of numerous things on earth and in heaven who creates and will annihilate them again is:

Infinite, he has no beginning and no end, Omnipresent.

He is one having no partner.

He does not resemble to any one of His creatures. And we will be having full faith in God who needs nothing in this world. Again we believe that
God is alive, he sees, he knows, he hears and he listens.

In short, there is nothing secret and unknown to Him. He is aware of every thing. God can do what he wishes. No one can interfere with His business,
and stand against His order. Neither has he given birth to nor been given birth by anyone. Nobody is His collaborator, partner and resembles to Him. He possesses endless and vast omnipotence. There is nothing on the surface of the earth which He could not do and will not be able to do in future. He talks and speaks. He has created every thing just from nothing.

لاَ إِلَٰهَ إِلَّا اللَّهُ مُحَمَّدٌ رَسُولُ اللَّهِ

**La Ilaahe illallah Muhammedun Rasulullah**
(Transliteration)

I bear witness that there is no deity save Allah (God), and that Muhammed is His servant and His Apostle. (translation)

Everybody who says the above blessed sentence with his word of mouth and certifies it with his

heart, is called Mü‘min (Believer). A man who wants to embrace Islam and follow its teachings can join the religion of Islam by saying this blessed
sentence. This is named as brief faith (Idjmaly iman). These two sentences contain every thing necessary and important to be believed and this is
the first step of faith. For the continuation of the religion’s path, this brief faith and this statement do not suffice. The Islamic faith means to believe more strongly and ardently; there are some principles and bases of faith.

Prophet Muhammed (peace be upon hiin) has defined the faith in these words :

Amentü billahi ve melâiketihi,
ve kütübihî ve rusülihî ve'l yevmi'l-âhıri
ve bi'l-kaderi, hayrihî ve şerrihi mina'llâhi teâlâ
ve'l-ba'sü ba'de'l mevt.
Eşhedü en lâ ilâhe illAllâh ve eşhedü enne
Muhammeden abdühû ve rasûlühü.
(Transliteration)

I Believe in God, in His angels, in His books in His prophets, in the day of judgement, and that good and evil fate comes from God and in resurrec-
tion after death. I bear witness that there is no deity save Allah (God) and Muhammed is His servant and His apostle.

**Belief (Iman)** : 

1. to believe that there is no deity save God, and Muhammed (Peace be upon him) is His servant and apostle, and to believe :
2. in the angels of God,
3. in the holy books of God,
4. in all the prophets of God,
5. in resurrection after death,
6. that goodness and badness are created by God and take place according to His will.

These six principles are the roots of faith; the foundations and bases of Islamic religion.

This is the obligation of every matured and wise man to have full belief in them. That means it has been strictly ordered by God. And this is named as detailed Belief (tafsily Iman.)

To believe in God renders the spirit of a man eminent. A Muslim who believes in God with his heart, bound to Him,, can attain the highest position ever achieved by mankind. A faith like this keeps its believer away from evil and bad thoughts.

And in this way badness and wickednes can not have any effect on his heart and he refrains from these. If he keeps in mind that there is some supreme power which controls and observes all his deeds and knows every thing which comes to his mind he will realise that one day he will reply to God for each of them. So he tries to keep his hands, tongue and heart far from every kind of badness. Even when nobody is seeing him, he is aware not doing any thing which may violate the moral qualities and be against the orders of God.

He can not be misled by anything and can not lose his right way. He decorates his spirit and his heart with every type of knowledge useful both for this world and the next world. In this way he protects himself from falsehood, insincerity and ignorance He loves not only the creator but also the created ones with all the zeal and heart, because they have also been created like him by God. He knows that supreme God has created all the human beings bestowing upon them equal rights. Therefore he does not feel proud by just taking himself as superior to those. Also he does not consider himself inferior to any one. He does not bow before anybody except God. He performs perfectly all the duties and moral obligations to his creator and created ones. In this way he gathers in him all the good habits and characters which are suitable both for humanbeings and humanity, thus proving himself as useful creature for other fellow-beings and at the same time he becomes a perfect believer (Mü’min). A man who comes to this world with spotless and clean spirit, goes to his God happily after developing all his aptitudes and faculties with the help of faith light and fulfilling all the duties towards humanity.

**MUSLIMS AND ISLAMIC FAITH**

**Islamic religion** means to believe in the orders and teachings of God and His prophets and be submissive and obedient to God; internally and externally to be pure and sincere.

**Muslim** is that person who submits himself to orders and teachings of God and His prophet.

Frequently the words Mü’min (believer) and Muslim have the same meaning and significance.

**RELIGION**

In case when a man is able to find his creator by means of his wisdom and thoughts, he cannot perceive even the reasons why somethings are allowed or forbidden and what God wishes. That is why God has informed us through his prophets about what is good and what is bad. And He has taught the humanbeings how they can embrace happiness both in this world and in the next world.

All these orders and prohibtions of God are given the name of Religion. The name of our religion is Islam religion or Muslim religion, and this name
was given by God Himself. Therefore, Muslim religion is full of wisdom and knowledge. It.is the religion of moral character, faith, peace and order. It is a faith for the survival of its believers.

Islamic faith has improved the right faiths of previous times and then redressed the falsifications introduced to those religions by evil doers thus
separating, right from wrong, also liberating the humanity from deep pits of evils and elevating it to that height and extreme of character and morale which is even unthinkable by human mind.
