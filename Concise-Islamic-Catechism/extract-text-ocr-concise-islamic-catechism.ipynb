# Extracting the Text from a PDF File

```python
import pytesseract
from pdf2image import convert_from_path

# Path to the PDF file
pdf_path = "Concise-Islamic-Catechism.pdf"

# Convert PDF pages to images
pages = convert_from_path(pdf_path)

# Initialize an empty string to store all extracted text
all_text = ""

# Loop through each page
for i, page in enumerate(pages):
    # Save the image
    image_path = f"out/page_{i + 1}.jpg"
    page.save(image_path, "JPEG")

    # Extract text from the image
    extracted_text = pytesseract.image_to_string(page, lang="eng")
    all_text += f"Page {i + 1}:\n{extracted_text}\n"

# Write the extracted text to a text file
output_file = "Concise-Islamic-Catechism.txt"
with open(output_file, "w") as f:
    f.write(all_text)

print(f"Extracted text saved to {output_file}")
```

## FIRST CHAPTER

### BASES OF FAITH IN ISLAM


Islam is established on two principal bases of faith :

1. There is no other God worth worshipping except Allah,
2. That Muhammed (peace be upon him) is the prophet sent by God to all the humanbeings.

The religion of Islam has summed up as a printiple in this sentence :

لاَ إِلَٰهَ إِلَّا اللَّهُ مُحَمَّدٌ رَسُولُ اللَّهِ

**La Ilaahe illallah Muhammedun Rasulullah**

(Transliteration)

"There is no deity save Allah (God) and Muhammed is His servant and His apostle". (Translation).

The man, who confirms this holy sentence by word of mouth and certifies it with his heart, is considered as having entered the religion of Islam.

This Faith is called Iman (Belief) and one who believes it, is named as Mü’min (Believer).

Let us have a glimpse of this endless existing universe. The world we live in, is one amongst million parts of this universe. So far as human being is concerned, he is like a minute particle as compared with the greatness of the world. Structurally weak and puny, the man is however the most honourable and the highest creature of this existing world. No other creature except man could attain this eminence. The man who has attained this height and honour, has equivalent responsibilities at the same time.

Therefore it is necessary for every wise man to think about : 

From where he has come,
Why he has come,
And where he will go,

What do we mean by this world? How did heaven and this universe come into existence and how did they acquire the present shape?

And man is forced by his mind to find answers to these and the relevant questions. Everybody who is not sick and possesses a sound mind, asks
himself questions of this kind and by thinking, he is able to find the pertaining answers. Just starting from ourselves let us also think about the earth, heaven and the inhabitants and creatures present in them.

This world, endless and limitless skies with millions and billions of stars, suns and moons; as from when are they existing and until when will
they continue to exist? Was there any time when they were not existing and will the time come when they will be no more existing? Coming into this
world, growing, dying, moving about, coming to a halt, completion and breaking into small pieces, are all clear-cut indications that they come into existence later on. Present day knowledge which is based on experience and observation also accepts that this universe had come into existance at later stage.

The conclusion of our thought :

If it is so, then it becomes obvious that the earth, heaven and everything which is present on the earth and in heaven, were created later on.

Now let us have a look at the existing universe!

Every thing present in it can be divided into two principal sections :

1 — Living ones,
2 — Lifeless ones.

These living creations include human beings, animals and plants; and these lifeless ones include every thing not included in the first category. Humanbeing is at the top of the creations present in the first section. And man with the highest degree of wisdom, is dominant over animals, plants and matter. He flies in the skies and goes to the depths of the oceans, he can make the finest instruments and can make use of them, and he proves his superiority by bringing other creatures under his control by means of his wisdom.

Besides man’s superiority, not only one human-being but all humanity is not in a position to create even the most minute particle of sand out of nothing.

This universe has been created by God who does not belong or resemble to it.


RECOGNITION OF GOD

The creator of numerous things on earth and in heaven who creates and will annihilate them again is:

Infinite, he has no beginning and no end, Omnipresent.

He is one having no partner.

He does not resemble to any one of His creatures. And we will be having full faith in God who needs nothing in this world. Again we believe that
God is alive, he sees, he knows, he hears and he listens.

In short, there is nothing secret and unknown to Him. He is aware of every thing. God can do what he wishes. No one can interfere with His business,
and stand against His order. Neither has he given birth to nor been given birth by anyone. Nobody is His collaborator, partner and resembles to Him. He possesses endless and vast omnipotence. There is nothing on the surface of the earth which He could not do and will not be able to do in future. He talks and speaks. He has created every thing just from nothing.

**La Ilaahe illallah Muhammedun Rasulullah**

The taking place of all the things at the predestinated time in the prefixed and predestinated shape by God is called in Islamic terminology as «qada». 

God has given a man the right of selection. He is free to do what he wishes. Therefore he is responsible for every thing he does. If a humanbeing wants to do good or bad things God creats this deed. This is what we mean «good or evil takes place with the consent of God.

Man strives and achieves and God creates. The eternal knowledge of God of every thing is not an obstacle to doing every thing freely by us. God
knows how we will behave with our will. He has allowed us to do freely what we wish.

In every work man has to do all possible things and not to leave any thing to be done in order to have a result in every work; then he should leave
the rest to God, he must depend on God. This what : we mean by the word **resignation** (Tawakkul).

There is no place in religion for those who believe that every thing will happen according to our fate and who do not strive for the achievement of
their goals and leave and wait in laziness.

Our Religion is a religion of life and religion of working faithfully honestly and continously.

**FIVE FUNDAMENTAL PRINCIPLES OF ISLAM**

Our beloved Prophet has stated the following : «Islam has been established on five fundamental principles.» They are: (a) There is no deity save Allah (God) and that Muhammed (peace be upon him) is His servant and His Apostle, (b) Prayers (Salat) (c) Legal alms (Zakat) (d) Fasting and (e) Pilgrimage.

According to the sacred sayings of our holy Prophet, Islam has been established as said before on five fundamental principles. For a believer to be
a perfect Muslim, he should fulfil all those five conditions which are described and explained as follows :

(a) To believe voluntarily in the existence and unity of God and that Muhammed (peace be upon him) is His servant and His Apostle, confirming this
by a word of mouth:

[arabic] 

«Ash-hadu an- la ilaha ill illallahu wa ash-hadu anna Muhammaden Abduhu wa Rasuluh». (Transliteration)

«I bear witness that there is no deity save Allah (God) and that Muhammed (peace be upon him) is His servant and His Apostle». (Translation)

(b) To perform the five enjoined prayers every day which are considered as the pillar of the Islamic religion. 

(c)  Those who are supposed to be rich are obliged to pay the legal alms (Zakat) of all their possessions and properties once a year.

(d) To fast in the lunar month of Ramadan.

(e) To perform the holy Pilgrimage once in his life in case one is in good health and has enough money.

** LIABILITY AGE AND DIVINE RULES CONCERNING THE AFFAIRS OF HUMANBEINGS: **

In order to obey all Divine Orders and Prohibitions and be responsible for all of them, a man should possess the following qualities :

1 - He should be wise.

2 - He should have acquired maturity and adultness.
   
a) Mad people and children are not obliged to follow these Divine Orders and Prohibitions.

b) According to Islam, the time of maturiy for girls is between 9-15 years and for boys between 12-15 years.

c) It is the obligation of the parents to teach their children the teachings of Islam before they attain this age. 

The affairs and mode of life of every wise and matured Muslim from religion's point of view can be divided into three categories :

1 - Those affairs which our religion orders us to do.

2 — Those which are not allowed in our religion.

3 — Those affairs whose performance or non-perfomance depends upon the will and choice of the devotee.

**Some of the ordered obligations :**

1. These are the things which are to be done positively by every believer, for example things like performance of ablution, praying five times daily, fasting in the lunar month of Ramadan and paying the legal alms (Zakat) are called in Islamic terminology Fards. This type of orders must be excuted absolutely, Those who do not fulfill these obligations, will get their due punishment in the next world and those who do not believe in these Divine orders, go out of the Islamic religion.

Devotions like having ablution, praying, fasting in the lunar month of Ramadan, giving the legal alms (Zakat) by rich people and praying at funerals
for the dead, are from these kinds of obligations.

(a) Every Muslim is personally responsible for the performance of some of these obligations (Fards). No body else can perform these Fards in his place. This type of devotions are named.as Fard-ul-Ayin (e.g. ablution, praying... etc.).

(b) Some other devotions are such that their performance by some sections of Muslims will absolve the rest. This kind of devotions are called Fard-ul-kifaya (e.g. prayers at funerals for the dead).

2. Some other obligations which are not based on decisive evidences like Fards, are also to be performed and those people who de not follow them will meet their due punishments. This type of orders are called **Wadjibs** (necessary) (e.g. to pray on the morning of the first day of Eids (a) Eid-ul-Fitr (b) Eid-ul-Adha, to sacrifice a victim (animal) on the Eid-ul-Adha according to one’s means.)

3. Those devotions which are neither Fard nor Wadjib but practised frequently and left off and on by our holy Prophet, are called **Sunna**.

a) Those who practise Sunna get God’s reward and those who leave them without any apparent excuse will be deprived of the intercession of the holy Prophet.

b) All the things which our beloved Prophet has himself continuously done or has advised us to do them, are named «Muekked Sunna» (confirmed Sunna) e.g. two raka’at before two Fards in the morning prayer, four raka’at before and two raka’at after the four Fards in the noon prayer, two raka’at after — four Fards in the night prayer (For details, please refer to the prayer’s section.).

c) The Sunna which were sometimes prayed and sometimes not prayed by our holy Prophet are termed as Ghayr-Muekked Sunna (unconfirmed Sunna).

The performance of this type of Sunna brings God’s reward and those who do ‘not perform them are not supposed to commit any sin. Four raka’at
Sunna before four Fards both in the after-noon and night prayers belong to this type. 

**Forbidden deeds :**

These are the deeds decisively forbidden by the Islamic religion like, murdering adultery, theft, lying disobeying the parents and conspiring. It is absolutely necessary not to commit these forbidden deeds. Those who commit them will be punished in the next world. The non-believers will go out of the Islamic religion.

The prohibited deeds with undecisive arguments and proofs are called **Makruh** (disagreeable acts).

The disagreeable deeds’ ( Makruh) are of two kinds :

(a) Inviolable (Tahrimen)

(b) Absolute (Tenzthen)

Those in the first category are near to the forbidden deeds. For example, like leaving the Wadjib part of the devotion. One should abstain from them.
Those things which disturb the devotions, are termed Disturbers (Mufsid) like speaking in prayers; eating and drinking while fasting.

Those things which can be done or left freely are named permissible (Mubah) like eating, drinking and walking ... etc
